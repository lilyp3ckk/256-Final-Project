```python
#these are all variables that will need inputs (when there is a p in front it means preferred)
#I put in answers to get the code running (I think we might have to title all answers)
try:
    user_g = input("What is your gender? (Male or Female) ")
    user_pg = input("Which gender are you attracted to? ")
    user_eye = input("What color are your eyes? ")
    user_peye = input("What color eyes do you prefer in a partner? ")
    user_hair = input("What color is your hair? ")
    user_phair = input("What color hair do you prefer in a partner? ")
    user_tattoos = input("Do you have tattoos? ")
    user_ptattoos = input("Do you like when your partner has tattoos? ")
    user_piercings = input("Do you have any piercings?")
    user_ppiercings = input("Do you like when your partner has piercings? ")
    user_act = input("What is your favorite activity? ")
    user_annoy = input("What is your biggest annoyance? ")
    user_music = input("What type of music do you like? ")
    user_movie = input("What is your favorite movie? ")
    user_kids = input("Do you want to have kids? ")
    user_color = input("What is your favorite color? ")


    users_dict = users.to_dict(orient='records')

    for user in users_dict:
        if user['Preferred Gender'] == user_g and user['Gender'] == user_pg:
            user['Count'] = user['Count'] + 1
            prefgender_df = pd.DataFrame(users_dict)
            prefgender_df = prefgender_df[prefgender_df['Count'] == 1]

    prefgender_dict = prefgender_df.to_dict(orient='records')        
    for user in prefgender_dict:        
        if user['Preferred Eye Color'] == user_eye:
            user['Count'] = user['Count'] + 1
        if user['Eye Color'] == user_peye:
            user['Count'] = user['Count'] + 1
        if user['Preferred Hair Color'] == user_hair:
            user['Count'] = user['Count'] + 1
        if user['Hair Color'] == user_phair:
            user['Count'] = user['Count'] + 1
        if user['Preferred Tattoos'] == user_tattoos:
            user['Count'] = user['Count'] + 1
        if user['Tattoos'] == user_tattoos:
            user['Count'] = user['Count'] + 1
        if user['Preferred Piercings'] == user_piercings:
            user['Count'] = user['Count'] + 1
        if user['Piercings'] == user_ppiercings:
            user['Count'] = user['Count'] + 1
        if user['Favorite Activity'] == user_act:
            user['Count'] = user['Count'] + 1
        if user['Music'] == user_music:
            user['Count'] = user['Count'] + 1
        if user['Annoyances'] == user_annoy:
            user['Count'] = user['Count'] + 1
        if user['Movie'] == user_movie:
            user['Count'] = user['Count'] + 1
        if user['Kids'] == user_kids:
            user['Count'] = user['Count'] + 1
        if user['Colors'] == user_color:
            user['Count'] = user['Count'] + 1
            
except NameError:
    print("Please ensure your typing is correct")
```
